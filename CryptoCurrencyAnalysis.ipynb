```
# New daaframe that hold ttthe closing price of all 3 crypto currencys
df = pd.DataFrame({
    'BTC':df_btc['Closing Price (USD)'],
    'ETH':df_eth['Closing Price (USD)'],
    'LTC':df_ltc['Closing Price (USD)']
})
```

```
df.tail()
```

```
# Get stats
df.describe()
```

```
# Getting visual
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

my_crypto = df
plt.figure(figsize = (12.2, 4.5))
for c in my_crypto.columns.values:
  plt.plot(my_crypto[c], label = c)

plt.title('Crypto Analysis')
plt.xlabel('Days')
plt.ylabel('Price in Dollars')
plt.legend()  
```

```
#Scale the data
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler(feature_range = (0, 100))
scaled = min_max_scaler.fit_transform(df)
scaled
```

```
df_scaled = pd.DataFrame(scaled, columns = df.columns)
df_scaled
```

```
# Visualize the scaled d
my_crypto = df_scaled
plt.figure(figsize = (12.4, 4.5))
for c in my_crypto.columns.values:
  plt.plot(my_crypto[c], label = c)

plt.title("Crypto Scaled")
plt.xlabel("Days")
plt.ylabel("Price in $")
plt.legend()
plt.show()
```

```
# Get the daily simple return

DSR = df.pct_change(1)
DSR
```

```
# Visualize the DSR

plt.figure(figsize = (12.2, 4.5))

for c in DSR.columns.values:
  plt.plot(DSR.index, DSR[c], label = c, lw=2, alpha =0.75)

plt.title("DSR")
plt.ylabel("PCT in decimal form")
plt.xlabel("Days")
plt.legend()
plt.show()
```

```
# Get the volatility
print('Crypto currency volatilty: ')
DSR.std()
```

```
DSR.mean()
```

```
# Get the corr matrix
DSR.corr()
```

```
#Let visualize corr
import seaborn as sns
plt.subplots(figsize = (4, 4))
sns.heatmap(DSR.corr(), annot = True, fmt='.2%')
```

```
DCSR = (DSR+1).cumprod()

DCSR
```

```
#Visualize the DCSR
plt.figure(figsize = (12.2, 4.5))
for c in DCSR.columns.values:
  plt.plot(DCSR.index, DCSR[c], label = c, lw=2, alpha=0.75)

plt.title("DCSR")
plt.xlabel("Day")
plt.ylabel("Growth of one dollar investment")
plt.legend()
plt.show()
```
